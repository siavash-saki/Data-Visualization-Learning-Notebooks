## Example

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
tips= sns.load_dataset('tips')
````

````python
tips.head()
````

### Scatter

````python
tips.iplot(kind='scatter',x='total_bill',y='tip',mode='markers',title='BILL vs TOTAL BILL',
           xTitle='Total Bill (USD)',yTitle='Tipp ($)',colors='blue',size=6)
````

### Bar

````python
tips.iplot(kind='bar',x='day',y='total_bill')
````

````python
tips.sum()[['total_bill','tip']].iplot(kind='bar')
````

### Box

````python
tips[['total_bill','tip']].iplot(kind='box')
````

### 3D Surface

````python
df = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df.iplot(kind='surface',colorscale='rdylbu')
````

### Spread

````python
tips[['total_bill','tip']].iplot(kind='spread')
````

### Bubble

````python
tips.iplot(kind='bubble',x='total_bill',y='tip',size='size')
````

### Histogram

````python
tips['total_bill'].iplot(kind='hist',bins=25)
````
